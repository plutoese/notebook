# 电影票房爬虫

```python
import pandas as pd
import requests

mdata = pd.read_excel("./data/movie_data.xlsx")
choosen_movie_id = list(mdata["MovieId"])

npdata = pd.read_excel("./data/movie_index.xlsx")
all_index = list(npdata["ID"])
```

```python
import random
from collections import deque

movie_detail_data = []
movie_indexes = set(choosen_movie_id)
num = 1
max_num = 2000
for i in range(100000):
    choice = random.choice(all_index)
    if choice not in movie_indexes:
        print(num)
        movie_indexes.add(choice)
        num += 1
        params = {"movieId": str(choice), "MethodName":"BoxOffice_GetMovieData_Details"}
        r = requests.post("http://www.endata.com.cn/API/GetData.ashx", data=params)
        movie_detail_data.append(r.json()["Data"]["Table"][0])
    
    if num > 1000:
        break
```

```python
movie_pd = pd.DataFrame(movie_detail_data)
movie_pd.to_excel("./data/movie_data4.xlsx", index=False)
```

```python
import pandas as pd

mdata = pd.read_excel("./data/movie_data.xlsx")
choosen_movie_id = list(mdata["MovieId"])
```

```python
import requests

movie_main_actors = dict()
for id in choosen_movie_id[4000:6000]:
    print(id)
    params = {"movieId": id, "MethodName":"BoxOffice_GetMovieData_Details_MainActor"}
    r = requests.post("http://www.endata.com.cn/API/GetData.ashx", data=params)
    movie_main_actors[id] = r.json()["Data"]["Table"]
```

```python
import pickle

F = open('movie_main_actors5.pkl', 'wb')
pickle.dump(movie_main_actors, F)
F.close()
```

```python
movie_main_actors = dict()

for i in range(1,6):
    F = open('movie_main_actors{}.pkl'.format(i), 'rb')
    E = pickle.load(F)
    movie_main_actors.update(E)

M = open('./data/movie_main_actors.pkl', 'wb')
pickle.dump(movie_main_actors, M)
M.close()
```
