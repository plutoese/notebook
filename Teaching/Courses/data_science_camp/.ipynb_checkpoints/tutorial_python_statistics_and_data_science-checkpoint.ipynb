### ndarray: NumPy库的心脏

- 整个NumPy库的基础是ndarray（N-dimensional array，N维数组）对象。它是一种由同质元素组成的多维数组。

- 同质指的是几乎所有元素的类型和大小都相同。

- 事实上，数据类型由另外一个叫做dtype（data-type，数据类型）的NumPy对象来指定：每个ndarray只有一种dtype类型。

````python
import numpy as np

# to create a ndarray
a = np.array([1, 2, 3])
print(a, " - ", type(a), ' - ', a.dtype)
print(a.shape)
````

````python
# to create a ndarray
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a, " - ", type(a), ' - ', a.dtype)
print(a.shape)
````

### 线性代数初步

#### 矩阵乘法

NumPy的线性代数中*是矩阵的逐元素乘积，而不是矩阵的点乘积。矩阵乘法使用dot函数。

````python
x = np.array([[1, 2, 3],[4, 5, 6]])
y = np.array([[6, 23], [-1, 7], [8, 9]])
x.dot(y)    # or np.dot(x, y)
````

#### 矩阵的逆

可用numpy.linalg的solve或者inv函数来求得矩阵的逆矩阵。

````python
from numpy.linalg import solve, inv

A = np.array([[2, 5], [-3, -7]])
I = np.eye(2)

A_inv = solve(A, I)
print(A_inv)

print(inv(A))
````

#### 行列式

可用numpy.linalg的det函数或者scipy.linalg.det来求得矩阵的行列式。

````python
from scipy.linalg import det

A = np.array([[1, 2], [3, 4]])
print(np.linalg.det(A))
print(det(A))
````

#### 特征根和特征向量

可用numpy.linalg的eig函数或者scipy.linalg.eig来求得矩阵的特征根和特征向量。

````python
A = np.array([[2, 3], [3, -6]])
e, v = np.linalg.eig(A)
print(e)
print(v)
````

## 从样本到样本统计量

### 掷骰子的实验

<div align=center>
<img src="https://i.imgur.com/GtbawRt.jpg" width = "70%" />
</div>

````python
import random

def coin_trial():
    heads = 0
    for i in range(100):
        if random.random() <= 0.5:
            heads +=1
    return heads
    
def simulate(n):
    trials = []
    for i in range(n):
        trials.append(coin_trial())
    return(sum(trials)/n)

simulate(1000)
````

### $t$检验的实现

可以关注[SciPy](https://docs.scipy.org/doc/scipy/reference/stats.html#module-scipy.stats)，这里有很多统计检验。

- 先看一个模拟数据的案例

````python
import random
import scipy.stats
random.seed(20)

N = 30
a = [random.gauss(50,5) for x in range(N)] # take N samples from population A
popmean = 50.5  # hypothesized population mean

tStat, pValue =  scipy.stats.ttest_1samp(a, popmean, axis=0)
print("P-Value:{0} T-Statistic:{1}".format(pValue, tStat)) # print the P-Value and the T-Statistic
````

````python
# calculation by hand
tvalue = (np.mean(a)-popmean)/(np.std(a)/np.sqrt(N-1))
pvalue = 2*(1 - scipy.stats.t.cdf(tvalue, N-1))
print("P-Value:{0} T-Statistic:{1}".format(pvalue, tvalue))
````

- 检验男性和女性的身高是否相等

````python
female_height = rdata[rdata["gender"]=="女"]["height"]
male_height = rdata[rdata["gender"]=="男"]["height"]

tStat, pValue = scipy.stats.ttest_ind(male_height, female_height, equal_var = False) #run independent sample T-Test
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) #print the P-Value and the T-Statistic
````

### 幸福的人生如何书写？- 一个简单的决策树分析

在这里我们用到了[scikit-learn](https://scikit-learn.org/stable/)，它可能是Python中处理机器学习问题最常用的库。它包括我们用过的所有模型以及很多我们没用过的模型。

#### 决策树

决策树通过树结构来表示各种可能的决策路径以及每个路径的结果。

<br>

**分类指标**

- 基尼系数 $G = \sum_{x \in X} p(x) (1 - p(x) )$

- 熵 $H = -\sum_{x \in X} p(x) \log _{2} p(x)$

<br>

**信息增益**

$$\operatorname{Gain}(S, A)= Entropy(S) - \sum_{v \in V a l u e s(A)} \frac{\left|S_{v}\right|}{|S|} \operatorname{Entropy}\left(S_{v}\right)$$

在每个结点，根据信息增益原则选择

- 哪个变量（属性）
- 变量（属性）的分割值

<br>

**案例**

<div align=center>
<img src="decision_tree04.png" width = "30%" />
</div>

因此，分类前

$$H(10 / 20,10 / 20)=-10 / 20 \log (10 / 20)-10 / 20 \log (10 / 20)=1$$

每个类的熵

$$H(home) =-6 / 12 \log (6 / 12)-6 / 12 \log (6 / 12)=1$$

$$H(away)=-4 / 8 \log (6 / 8)-4 / 8 \log (4 / 8)=1$$

分类后的熵为

$$12 / 20^{*} \mathrm{H}(\text { home })+8 / 20^{*} \mathrm{H}(\text { away })=1$$

<br>

<div align=center>
<img src="decision_tree05.png" width = "30%" />
</div>

每个类的熵

$$H(5pm)=-1 / 4 \log (1 / 4)-3 / 4 \log (3 / 4)$$

$$H(7pm)=-9 / 12 \log (9 / 12)-3 / 12 \log (3 / 12)$$

$$H(9pm)=-0 / 4 \log (0 / 4)-4 / 4 \log (4 / 4)=0$$

分类后的熵为

$$4 / 20^{*} \mathrm{H}(1 / 4,3 / 4)+12 / 20^{*} \mathrm{H}(9 / 12,3 / 12)+4 / 20^{*} \mathrm{H}(0 / 4,4 / 4)=0.65$$

因此信息增益为

$$1-0.65=0.35$$

所以，毫无疑问应该选择后者。每个节点都选择信息增益最大的属性进行叉裂。

````python
import re
import graphviz 
from sklearn import tree

rdata["job"] = 0
rdata.loc[rdata["employment"]=="是","job"]=1

rdata["married"] = 0
rdata.loc[rdata["marriage"]=="初婚有配偶","married"]=1

rdata["city"] = 1
rdata.loc[rdata["urban"]=="农村","city"]=0

rdata["female"] = 1
rdata.loc[rdata["gender"]=="男","female"]=0

rdata["house"] = 0
rdata.loc[rdata["house_owned_by_myself"]=="是","house"]=1

rdata["class"] = rdata["social_class"].apply(lambda x: int(re.split("\(",x)[0]))

rdata["happy"] = 0
rdata.loc[rdata["happiness"]=="比较不幸福","happy"]=1
rdata.loc[rdata["happiness"]=="说不上幸福不幸福","happy"]=2
rdata.loc[rdata["happiness"]=="比较幸福","happy"]=3
rdata.loc[rdata["happiness"]=="非常幸福","happy"]=4
happy_label = ["非常不幸福", "比较不幸福",  "说不上幸福不幸福", "比较幸福", "非常幸福"]

rdata["happy2"] = 0
rdata.loc[rdata["happiness"]=="比较幸福","happy2"]=1
rdata.loc[rdata["happiness"]=="非常幸福","happy2"]=1
happy2_label = ["不幸福", "幸福"]

rdata["healthy"] = 0
rdata.loc[rdata["health"]=="比较不健康","healthy"]=1
rdata.loc[rdata["health"]=="一般","healthy"]=2
rdata.loc[rdata["health"]=="比较健康","healthy"]=3
rdata.loc[rdata["health"]=="很健康","healthy"]=4

columns = ["city","female","educ","height","weight","income","famincome","birthyear","healthy","house","class","job","married","spouse_educ"]

X = rdata[columns]
y = rdata["happy"]
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, y)
````

````python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=columns,  class_names=happy_label,  filled=True, rounded=True,  special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
````

````python
columns = ["happy2","city","female","educ","height","weight","income","famincome","birthyear","healthy","house","class","job","married","spouse_educ"]
data = rdata[columns]
train_data = data.iloc[0:6800,]
test_data = data.iloc[6800:,]

X = train_data[columns[1:]]
y = train_data[columns[0]]
clf = tree.DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X, y)

X_test = test_data[columns[1:]]
y_test = test_data[columns[0]]
y_pred=clf.predict(X_test)

score = clf.score(X_test, y_test)
print(score)
````

````python
for depth in range(2, 30):
    clf = tree.DecisionTreeClassifier(max_depth=depth)
    clf = clf.fit(X, y)

    X_test = test_data[columns[1:]]
    y_test = test_data[columns[0]]
    y_pred=clf.predict(X_test)

    score = clf.score(X_test, y_test)
    print("Max Depth:{} - Score:{}".format(depth,score))
````

- SVM分类器
