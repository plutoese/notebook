# 机器学习与经济学——简述

## 大数据、机器学习与人工智能

- 大数据 —— 巨量数据、高维度数据、非结构性数据（文本、音频、图像等）

<br>

- 机器学习指的是从数据中识别出规律并以此完成预测、分类及聚类等任务的算法总称。（王芳等，2020）
  - 监督学习 
  - 非监督学习 - 垃圾邮件处理

<br>
 
- 机器学习（尤其是深度学习）是人工智能的一个组成部分（Taddy，2018）

## 机器学习在经济学中的应用

### 图像数据

- 卫星遥感数据 —— 城市的扩张

<br>

- 谷歌街道图像数据
  - Glaeser等（2018）利用2007-2014年鉴12200张纽约和3608张波士顿的街景图像，利用机器学习中的视觉算法，识别出图像中建设的物理属性（高度等），并与美国社区调查（ACS）数据结合，研究纽约、波士顿地区的房价和贫富差距问题。

<br>

- Edelman等（2017）通过机器学习判别Airbnb上的用户头像性别进而分析租房平台上是否存在性别歧视。（人种歧视？）

### 文本数据

- Hansen等（2018）研究透明度政策如何影响政府内决策过程。其背景是美国联邦公开市场委员会在1993年通过决议公开了内部会议的发言记录。作者将该政策视作自然实验，观测委员会成员的发言内容在该年前后的变化。研究基于的文本信息包含5万多次发言，总计500多万个单词。利用LDA模型，从这些海量文本中提取40个不同的主题。任意一个成员的每一条发言都可以对应到这些主题中的一个或几个上。这样，每个成员发言中各个主题的占比及成员间发言的相似度等指标就可以被计算出来，作者就可以使用OLS检验透明度政策对这些被解释变量的影响了。

<br>

- Iaria等（2018）试图研究一战冲击是否会影响跨国学术交流合作。作者预期战争会降低论文相似程度，基于如下逻辑：战争会使得跨国学术合作被迫中断，这将导致同盟国和协约过各自的论文标题相似度下降。其中关键要解决的问题是比较论文间的相似度：样本包括4万多篇论文。作者采用语义分析来比较论文标题间的相似程度，发现论文相似度在一战爆发后显著下降。

### 预测

**机器学习与计量经济学**

- 在方法-偏差权衡中，计量经济学方法宁愿付出方差较大的代价，也不能放弃无偏这一性质（Athey，2019）

- 而机器学习的目的是进行预测，因此它选择用偏差来换取更小的方法以提高预测性能

- 计量经济学的目标是“可解释”，因此函数形式越简单越好

- 机器学习恰好相反，只要函数能很好地进行预测，函数形式再复杂都没关系

<br>

**预测应用**

- Chalfin等（2016）采用决策树等预测警察的工作质量，作者通过警员入职申请中提及的社会经济状况、婚姻、是否服役等信息，预测其未来工作中是否偏向使用暴力。

- 刘涛熊和徐晓飞（2015）通过互联网搜索数据对GDP进行预测。

### 机器学习与因果推断

- DID与机器学习相结合
  - Cicala（2017）的研究试图评估美国电网从国家计划发电到市场自动调整发电量这一措施所带来的的收益。由于政府计划发电的废除是逐年逐地区推进的，因此可以使用一个典型的DID框架来分析市场介入带来的因果效应。但是平行趋势假设较难满足。因此，他利用历史数据构建反事实的问题，将燃料价格波动纳入考量，采用了随机森林的方法预测了每个地区若没有市场介入时发电量的反事实。
  
<br>
  
- RD设计
  - 例如考试分数（例如60分）是能否上大学的分界线。RD近似地认为59分和60分的学生同质，从而可以进行RD估计。即使如此，烤59分和60分的学生在能力上相同的假设可能太强。机器学习可以帮助研究者消除这一差异（Varian，2016）。例如通过预测得到小于59分的样本的工资预测，作为反事实。Imbens和Wager（2019）利用凸优化的数值方法进行断点回归的因果推断，超越了传统的局部线性回归。

<br>
  
- 工具变量方法
  - Angrist和Krueger（1991）的经典论文，其中在第一阶段工具变量的相关性问题，本质上是外生变量对内生教育变量的预测能力，这正是机器学习最擅长的地方。因此有作者（Belloni等，2012）采用正则化回归，例如LASSO等，也有学者（Hartford等，2017）采用神经网络等非线性方法来进行第一阶段的估计。
